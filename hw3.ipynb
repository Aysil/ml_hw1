Судя по экспериментам выше, посимвольный частотный метод позволяет довольно неплохо определять пробелы, будем это использовать для начального приближения метода с сэмплированием, кроме того будем для "затравки" алгоритма использовать одно- и двухбуквенные слова из ограниченного списка возможных слов

```python
one_letter_words = ['ж', 'в', 'и', 'у', 'а', 'с', 'к', 'я', 'о']
# кроме ее
two_letter_words = ['то', 'за', 'он', 'ко', 'та', 'во', 'же', 'ел', 'ух', 'те', 'но', 'их', 'из', 'со', 'ну', 'от',
'ус', 'ха', 'эй', 'эх', 'ты', 'ли', 'аж', 'ею', 'ни', 'уж', 'ай', 'не', 'ей', 'бы', 'гм', 'вы', 'мы', 'об',
'на', 'ах', 'до', 'ох', 'ту', 'ум', 'им', 'да', 'по']
```

```python
def sample_one_letter_word(skip=[]):
    filtered = [e for e in one_letter_words if e not in skip]
    if len(filtered) == 0:
        return None
    return filtered[random.randint(0, len(filtered) - 1)]
```

```python
def sample_two_letter_word(known_letters='**', skip=[]):
    filtered = two_letter_words.copy()
    if known_letters[0] != '*':
        filtered = [e for e in filtered if e[0] == known_letters[0]]
    if known_letters[1] != '*':
        filtered = [e for e in filtered if e[1] == known_letters[1]]
    skip_full = [e for e in skip if e not in [[known_letters[0], known_letters[1]]]]
    filtered = [e for e in filtered if e[0] not in skip_full and e[1] not in skip_full]
    if len(filtered) == 0:
        return None
    return filtered[random.randint(0, len(filtered) - 1)]
```

```python
def translation(text, known_letters={}):
    t = ''
    for l in text:
        if l in known_letters:
            t += known_letters[l].upper()
        else:
            t += l
    return t
```

```python
def get_freq_bigrams(text, known_letters={}):
    bigrams_curr = {}
    for idx in range(len(text)):
        l = text[idx]
        if l in known_letters:
            continue
        if idx > 0:
            if text[idx - 1] in known_letters:
                if l not in bigrams_curr:
                    bigrams_curr[l] = {'count': 0, 'bigrams': {}}
                tr_l = known_letters[text[idx - 1]]
                if f'{tr_l}*' not in bigrams_curr[l]['bigrams']:
                    bigrams_curr[l]['bigrams'][f'{tr_l}*'] = 0
                bigrams_curr[l]['bigrams'][f'{tr_l}*'] += 1
                bigrams_curr[l]['count'] += 1
        if idx < len(text) - 1:
            if text[idx + 1] in known_letters:
                if l not in bigrams_curr:
                    bigrams_curr[l] = {'count': 0, 'bigrams': {}}
                tr_l = known_letters[text[idx + 1]]
                if f'*{tr_l}' not in bigrams_curr[l]['bigrams']:
                    bigrams_curr[l]['bigrams'][f'*{tr_l}'] = 0
                bigrams_curr[l]['bigrams'][f'*{tr_l}'] += 1
                bigrams_curr[l]['count'] += 1
    counts = [bigrams_curr[l]['count'] for l in bigrams_curr]
    if len(counts) == 0:
        return (None, None)
    k = list(bigrams_curr.keys())[counts.index(max(counts))]
    return (k, bigrams_curr[k]['bigrams'])
```

```python
def build_probabilities(bigrams: {}, known_letters: {}):
    current_letters = [l for l in letters_list if l not in list(known_letters.values())]
    probs = {}
    for l in current_letters:
        probs[l] = 1.0
        for bigram in bigrams:
            if bigram[0] == '*':
                probs[l] *= (bigram_freq[f'{l}{bigram[1]}'] ** bigrams[bigram])
            else:
                probs[l] *= (bigram_freq[f'{bigram[0]}{l}'] ** bigrams[bigram])
    s = sum(list(probs.values()))
    if s == 0:
        return {k: 1 / len(probs) for k in probs}
    return {k: probs[k] / s for k in probs}
```

```python
def sample_with_probs(probs):
    v = random.random()
    curr_v = 0.0
    i = 0
    freqs = list(probs.values())
    while i < len(freqs) and curr_v < v:
        curr_v += freqs[i]
        i += 1
    return list(probs.keys())[i - 1]
```

```python
def sample_translation(text, known_letters={}):
    words_curr = [e.strip() for e in text.split(' ')]
    one_letter_words_curr = [w for w in words_curr if len(w) == 1 and w not in known_letters]
    for l in one_letter_words_curr:
        if l in known_letters:
            continue
        new_l = sample_one_letter_word(skip=list(known_letters.values()))
        if new_l is None:
            continue
        known_letters[l] = new_l
        
    two_letter_words_curr = [w for w in words_curr if len(w) == 2 and (w[0] not in known_letters or w[1] not in known_letters)]
    for w in two_letter_words_curr:
        if w[0] == w[1]:
            known_letters[w[0]] = 'е'
        else:
            if w[0] in known_letters and w[1] in known_letters:
                continue
            kl = f'{w[0] if w[0] in known_letters else "*"}{w[1] if w[1] in known_letters else "*"}'
            new_w = sample_two_letter_word(known_letters=kl, skip=list(known_letters.values()))
            if new_w is None:
                if w[0] in known_letters:
                    return w[0]
                else:
                    return w[1]
            known_letters[w[0]] = new_w[0]
            known_letters[w[1]] = new_w[1]
    
    letter, bigrams = get_freq_bigrams(text, known_letters=known_letters)
    while letter is not None:
        probs = build_probabilities(bigrams=bigrams, known_letters=known_letters)
        new_l = sample_with_probs(probs)
        known_letters[letter] = new_l
        letter, bigrams = get_freq_bigrams(text, known_letters=known_letters)
    return translation(text, known_letters)
```

```python
def find_least_used(words_found, letters):
    counts = {l: sum([len(e) for e in words_found if l in e]) for l in letters}
    if 0 in list(counts.values()):
        idx = list(counts.values()).index(0)
        return list(counts.keys())[idx]
    counts = {l: 1.0 / counts[l] for l in counts}
    s = sum(list(counts.values()))
    counts = {l: counts[l] / s for l in counts}
    if len(counts) == 0:
        return None
    return sample_with_probs(counts)
```

На вход методу для перевода подается строка для перевода, словарь букв, для которых мы полагаем на данный момент перевод известным и словарь реально существующих слов. Последовательность действия алгоритма сводится к следующему:
- находим букву, которая входит в максимальное число биграмм с уже известными буквами
- строим вероятностное распределение для этой буквы на основе частот биграмм построенному по "Войне и миру"
- сэмплируем из этого распределения букву
- повторяем процедуру до получения перевода всего текста
- проходим по получившемуся переводу и выделяем реально существующие слова по переданному словарю
- для реально существующих слов сохраняем соответствия букв в словарь
- повторяем процедуру с использованием этого словаря в качестве начального приближения
- если в течении какого-то числа итераций не происходит изменений, то выбрасываем из словаря буквы, которые используются в минимальном числе "реальных" слов и повторяем процедуру с таким словарем в качестве инициализации

```python
def find_translation(text, known_letters={}, iters=100, vocab=[]):
    known_letters_best = known_letters.copy()
    attempts = 0
    words_found = []
    for _ in range(iters):
        known_letters = known_letters_best.copy()
        tr = sample_translation(text, known_letters=known_letters)
        if len(tr) == 1:
            words_found = [e for e in words_found if tr in e]
            if tr in known_letters_best:
                known_letters_best.pop(tr)
            continue
        words_curr = [e.strip() for e in tr.lower().split(' ')]
        words_curr = [e for e in words_curr if len(e) > 2 and e in vocab and e not in words_found]
        words_found += words_curr
        if len(words_curr) == 0:
            attempts += 1
            if attempts >= 10:
                k = find_least_used(words_found, list(known_letters_best.values()))
                if k is not None:
                    known_letters_inv = {v: k for k, v in known_letters.items()}
                    if k in known_letters_inv and known_letters_inv[k] in known_letters_best:
                        known_letters_best.pop(known_letters_inv[k])
                    words_found = [e for e in words_found if k in e]
                attempts = 0
            continue
        known_letters_inv = {v: k for k, v in known_letters.items()}
        for w in words_curr:
            for l in w:
                known_letters_best[known_letters_inv[l]] = l
    return (translation(text, known_letters_best), known_letters_best)
```

```python
def get_freq_ngrams(text, known_letters={}):
    ngrams_curr = {}
    for idx in range(len(text)):
        l = text[idx]
        if l in known_letters:
            continue
        if idx > 0:
            if text[idx - 1] in known_letters:
                if l not in ngrams_curr:
                    ngrams_curr[l] = {'count': 0, 'ngrams': {}}
                tr_l = known_letters[text[idx - 1]]
                if f'{tr_l}*' not in ngrams_curr[l]['ngrams']:
                    ngrams_curr[l]['ngrams'][f'{tr_l}*'] = 0
                ngrams_curr[l]['ngrams'][f'{tr_l}*'] += 1
                ngrams_curr[l]['count'] += 1
        if idx < len(text) - 1:
            if text[idx + 1] in known_letters:
                if l not in ngrams_curr:
                    ngrams_curr[l] = {'count': 0, 'ngrams': {}}
                tr_l = known_letters[text[idx + 1]]
                if f'*{tr_l}' not in ngrams_curr[l]['ngrams']:
                    ngrams_curr[l]['ngrams'][f'*{tr_l}'] = 0
                ngrams_curr[l]['ngrams'][f'*{tr_l}'] += 1
                ngrams_curr[l]['count'] += 1
        if idx > 0:
            if idx > 1:
                if text[idx - 2] in known_letters and text[idx - 1] in known_letters:
                    if l not in ngrams_curr:
                        ngrams_curr[l] = {'count': 0, 'ngrams': {}}
                    tr_l_1 = known_letters[text[idx - 1]]
                    tr_l_2 = known_letters[text[idx - 2]]
                    if tr_l_1 == ' ':
                        continue
                    if f'{tr_l_2}{tr_l_1}*' not in ngrams_curr[l]['ngrams']:
                        ngrams_curr[l]['ngrams'][f'{tr_l_2}{tr_l_1}*'] = 0
                    ngrams_curr[l]['ngrams'][f'{tr_l_2}{tr_l_1}*'] += 1
                    ngrams_curr[l]['count'] += 1
            if idx < len(text) - 1:
                if text[idx - 1] in known_letters and text[idx + 1] in known_letters:
                    if l not in ngrams_curr:
                        ngrams_curr[l] = {'count': 0, 'ngrams': {}}
                    tr_l_1 = known_letters[text[idx - 1]]
                    tr_l_2 = known_letters[text[idx + 1]]
                    if f'{tr_l_1}*{tr_l_2}' not in ngrams_curr[l]['ngrams']:
                        ngrams_curr[l]['ngrams'][f'{tr_l_1}*{tr_l_2}'] = 0
                    ngrams_curr[l]['ngrams'][f'{tr_l_1}*{tr_l_2}'] += 1
                    ngrams_curr[l]['count'] += 1
        if idx < len(text) - 2:
            if text[idx + 1] in known_letters and text[idx + 2] in known_letters:
                if l not in ngrams_curr:
                    ngrams_curr[l] = {'count': 0, 'ngrams': {}}
                tr_l_1 = known_letters[text[idx + 1]]
                tr_l_2 = known_letters[text[idx + 2]]
                if tr_l_1 == ' ':
                        continue
                if f'*{tr_l_1}{tr_l_2}' not in ngrams_curr[l]['ngrams']:
                    ngrams_curr[l]['ngrams'][f'*{tr_l_1}{tr_l_2}'] = 0
                ngrams_curr[l]['ngrams'][f'*{tr_l_1}{tr_l_2}'] += 1
                ngrams_curr[l]['count'] += 1
    counts = [ngrams_curr[l]['count'] for l in ngrams_curr]
    if len(counts) == 0:
        return (None, None)
    k = list(ngrams_curr.keys())[counts.index(max(counts))]
    return (k, ngrams_curr[k]['ngrams'])
```

```python
def build_probabilities(ngrams: {}, known_letters: {}):
    current_letters = [l for l in letters_list if l not in list(known_letters.values())]
    trigrams_present = len([e for e in list(ngrams.keys()) if len(e) == 3]) > 0
    probs = {}
    #print(ngrams)
    min_p_trigram = 1.5 * (10 ** (-6))
    for l in current_letters:
        probs[l] = 1.0
        for ngram in ngrams:
            if trigrams_present:
                if len(ngram) == 2:
                    continue
                if ngram[0] == '*':
                    probs[l] *= (max([trigram_freq[f'{l}{ngram[1]}{ngram[2]}'], min_p_trigram]) ** ngrams[ngram])
                elif ngram[1] == '*':
                    probs[l] *= (max([trigram_freq[f'{ngram[0]}{l}{ngram[2]}'], min_p_trigram]) ** ngrams[ngram])
                else:
                    probs[l] *= (max([trigram_freq[f'{ngram[0]}{ngram[1]}{l}'], min_p_trigram]) ** ngrams[ngram])
            else:
                if ngram[0] == '*':
                    probs[l] *= (bigram_freq[f'{l}{ngram[1]}'] ** ngrams[ngram])
                else:
                    probs[l] *= (bigram_freq[f'{ngram[0]}{l}'] ** ngrams[ngram])
    #print([e for e in list(ngrams.keys()) if len(e) == 3])
    #print(probs)
    s = sum(list(probs.values()))
    if s == 0:
        return {k: 1 / len(probs) for k in probs}
    return {k: probs[k] / s for k in probs}
```

```python
def sample_translation(text, known_letters={}):
    words_curr = [e.strip() for e in text.split(' ')]
    one_letter_words_curr = [w for w in words_curr if len(w) == 1 and w not in known_letters]
    for l in one_letter_words_curr:
        if l in known_letters:
            continue
        new_l = sample_one_letter_word(skip=list(known_letters.values()))
        if new_l is None:
            continue
        known_letters[l] = new_l
        
    two_letter_words_curr = [w for w in words_curr if len(w) == 2 and (w[0] not in known_letters or w[1] not in known_letters)]
    #print(two_letter_words_curr)
    for w in two_letter_words_curr:
        if w[0] == w[1]:
            known_letters[w[0]] = 'е'
        else:
            if w[0] in known_letters and w[1] in known_letters:
                continue
            kl = f'{w[0] if w[0] in known_letters else "*"}{w[1] if w[1] in known_letters else "*"}'
            #print(kl)
            #print(list(known_letters.values()))
            new_w = sample_two_letter_word(known_letters=kl, skip=list(known_letters.values()))
            #print(new_w)
            if new_w is None:
                if w[0] in known_letters:
                    return w[0]
                else:
                    return w[1]
            known_letters[w[0]] = new_w[0]
            known_letters[w[1]] = new_w[1]
    
    letter, ngrams = get_freq_ngrams(text, known_letters=known_letters)
    while letter is not None:
        #print(ngrams)
        probs = build_probabilities(ngrams=ngrams, known_letters=known_letters)
        #print(probs)
        new_l = sample_with_probs(probs)
        known_letters[letter] = new_l
        letter, ngrams = get_freq_ngrams(text, known_letters=known_letters)
    #print(text)
    return translation(text, known_letters)
```

```python
def find_translation(text, known_letters={}, iters=100, vocab=[]):
    known_letters_best = known_letters.copy()
    attempts = 0
    words_found = []
    for _ in range(iters):
        known_letters = known_letters_best.copy()
        tr = sample_translation(text, known_letters=known_letters)
        #print(tr)
        if len(tr) == 1:
            words_found = [e for e in words_found if tr in e]
            if tr in known_letters_best:
                known_letters_best.pop(tr)
            continue
        words_curr = [e.strip() for e in tr.lower().split(' ')]
        words_curr = [e for e in words_curr if len(e) > 2 and e in vocab and e not in words_found]
        words_found += words_curr
        if len(words_curr) == 0:
            attempts += 1
            if attempts >= 10:
                k = find_least_used(words_found, list(known_letters_best.values()))
                if k is not None:
                    known_letters_inv = {v: k for k, v in known_letters.items()}
                    if k in known_letters_inv and known_letters_inv[k] in known_letters_best:
                        known_letters_best.pop(known_letters_inv[k])
                    words_found = [e for e in words_found if k in e]
                attempts = 0
            continue
        known_letters_inv = {v: k for k, v in known_letters.items()}
        for w in words_curr:
            for l in w:
                known_letters_best[known_letters_inv[l]] = l
    return (translation(text, known_letters_best), known_letters_best)
```
